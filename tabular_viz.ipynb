```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='darkgrid')
```

```python
df=pd.read_csv('car data.csv')
df.head()
```

```python
cont = ['Selling_Price', 'Present_Price', 'Kms_Driven', 'Year']
cont2 = ['Selling_Price', 'Present_Price']
cat = [i for i in df.columns if i not in cont]
cat.remove('Car_Name')
```

# Categorical

```python
def count_categorical_plot(data, col_list):
    """
    Count plot of categorical variables 
    """
    plt.figure(figsize=(len(col_list)*2,len(col_list)*2))
    for i, el in enumerate(col_list, 1):
        plt.subplot(int(len(col_list)/2), int(len(col_list)/2), i)
        sns.countplot(x=data[el])
        plt.xticks(rotation=45)
        plt.xlabel(el, fontweight='bold', fontsize=12)
    plt.tight_layout()
```

```python
count_categorical_plot(df, cat)
```

# Continuous

```python
def central_tendency_plot(data, col_list):
    plt.figure(figsize=(len(col_list)*3,len(col_list)*2))
    for i, el in enumerate(col_list, 1):
        # plt.subplot(int(len(cont)/2), int(len(cont)/2), i)
        plt.subplot
        f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (.15, .85)})
        sns.boxplot(x=data[el], ax=ax_box)
        # sns.swarmplot(x=df[el], ax=ax_box)
        sns.histplot(x=data[el], ax=ax_hist, kde=True)
        plt.axvline(x=data[el].mean(), color='red', ls='--', label="Mean")
        plt.axvline(x=data[el].median(), color='green', ls='--', label='Median')
        plt.legend(loc='upper right')
        plt.xticks(rotation=45)
        ax_box.set(xlabel='')
        plt.xlabel(el, fontweight='bold', fontsize=12)
    plt.tight_layout()
```

```python
central_tendency_plot(df, cont)
```

```python
plt.figure(figsize=(len(cont)*3,len(cont)*2))
for i, el in enumerate(cont, 1):
    plt.subplot(int(len(cont)/2), int(len(cont)/2), i)
    sns.histplot(x=df[el], kde=True)
    plt.axvline(x=df[el].mean(), color='red', ls='--', label="Mean")
    plt.axvline(x=df[el].median(), color='green', ls='--', label='Median')
    plt.legend(loc='upper right')
    plt.xticks(rotation=45)
plt.tight_layout()
```
